```python
# Handpick important columns
df1_columns = ['job_title', 'full_name', 'linkedin_profile_id', 'linkedin_profile_url', 'location', 'sales_navigator_company_id',
              'tenure_length', 'company_id', 'company_name', "industry"]
```

```python
cleaned_df1 = df[df1_columns]
```

```python
cleaned_df1['company_id'].isnull
```

```python
cleaned_df1.loc[:, 'company_id'] = cleaned_df1['company_id'].fillna(0).astype(int).astype(str)
```

```python
cleaned_df1['company_id']
```

```python
# Extracting the metadata from the 'meta' column
df['meta'] = df['meta'].apply(ast.literal_eval)
```

```python
# Creating the columns using lambda function to each column of the metadata
metadata_cols = df['meta'].apply(lambda x: pd.Series(x))
```

```python
metadata_cols['updated_at']
```

```python
metadata_cols['sales_navigator_url']
```

```python
# Handpick the important columns from the metadata
metadata_cols.columns
```

```python
df2_columns = ['title', 'job_description', 'job_url', 'job_id', 'poster_full_name', 'poster_id', 'poster_profile_url',
              'company_id', 'number_employees']
```

```python
metadata_cols["poster_id"]
```

```python
cleaned_df2 = metadata_cols[df2_columns]
```

```python
cleaned_df2['company_id']
```

```python
merged_df = pd.merge(cleaned_df2, cleaned_df1, on="company_id")
```
